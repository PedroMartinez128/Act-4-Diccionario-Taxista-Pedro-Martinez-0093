```python
# Diccionario global
taxistas = {}

def agregar_Taxista():
    print("\n--- AGREGAR TAXISTA ---")
    unidad = input("Número de unidad: ")
    zona = input("Zona de operación: ")
    plataforma = input("Plataforma digital: ")
    marca = input("Marca del auto: ")
    calificacion = input("Calificación: ")
    turno = input("Turno: ")

    taxistas[unidad] = {
        "zona": zona,
        "plataforma": plataforma,
        "marca": marca,
        "calificacion": calificacion,
        "turno": turno
    }
    print("Taxista registrado.")

def Listar_Taxista():
    print("\n" + "="*95)
    # Encabezados con alineación: <10 significa 10 espacios a la izquierda
    print(f"{'UNIDAD':<10} | {'MARCA':<15} | {'ZONA':<15} | {'PLATAFORMA':<15} | {'CALIF':<10} | {'TURNO':<15}")
    print("-" * 95)

    for unidad, d in taxistas.items():
        print(f"{unidad:<10} | {d['marca']:<15} | {d['zona']:<15} | {d['plataforma']:<15} | {d['calificacion']:<10} | {d['turno']:<15}")
    print("="*95)

def buscar_Taxista():
    print("\n--- BUSCAR ---")
    u = input("Número de unidad: ")
    d = taxistas[u]
    print(f"\nDatos: Unidad {u}, Marca {d['marca']}, Zona {d['zona']}, Turno {d['turno']}")

def borrar_Taxista():
    print("\n--- BORRAR ---")
    u = input("Número de unidad a eliminar: ")
    del taxistas[u]
    print("Eliminado correctamente.")

def actualizar_Taxista():
    print("\n--- ACTUALIZAR ---")
    u = input("Número de unidad a modificar: ")
    taxistas[u]["zona"] = input("Nueva zona: ")
    taxistas[u]["plataforma"] = input("Nueva plataforma: ")
    taxistas[u]["marca"] = input("Nueva marca: ")
    taxistas[u]["calificacion"] = input("Nueva calificación: ")
    taxistas[u]["turno"] = input("Nuevo turno: ")
    print("Datos actualizados.")

# Menú principal
while True:
    print("\n1. Agregar Taxista")
    print("2. Listar Taxistas")
    print("3. Buscar Taxista")
    print("4. Borrar Taxista")
    print("5. Actualizar Taxista")
    print("6. Salir")

    opcion = input("\nSeleccione una opción: ")

    if opcion == "1":
        agregar_Taxista()
    elif opcion == "2":
        Listar_Taxista()
    elif opcion == "3":
        buscar_Taxista()
    elif opcion == "4":
        borrar_Taxista()
    elif opcion == "5":
        actualizar_Taxista()
    elif opcion == "6":
        print("Saliendo...")
        print("Programa realizado por Pedro Martinez NL = 34")
        break
```
